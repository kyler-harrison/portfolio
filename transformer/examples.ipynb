```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
embed_len = 100
num_words = 100
d_model = embed_len * num_words
```

```python
def gen_pos_encoding(pos, d_model, max_i):
    """
    generates sequence of positional encodings for a fixed value of pos
    """
    encodings = []
    indices = []
    
    for i in range(0, max_i):
        indices.append(i)
        
        if (i % 2 == 0):
            encodings.append(np.sin(pos / (10000**(2 * i / d_model))))
        else:
            encodings.append(np.cos(pos / (10000**(2 * i / d_model))))
            
    return (indices, encodings)
```

```python
i0_pos = []
poss = []

for pos in range(num_words):
    (indices, encodings) = gen_pos_encoding(pos, d_model, embed_len)
    i0_pos.append(encodings[0])
    poss.append(pos)
    if (pos % 10 == 0):
        
        plt.plot(indices, encodings)
        plt.title(f"pos={pos}")
        plt.xlabel("i")
        plt.ylabel("encoding")
        plt.show()
plt.plot(poss, i0_pos)
```
